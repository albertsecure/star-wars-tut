```python
# from terminusdb_client import Client
from terminusdb_client import WOQLClient
from terminusdb_client.woqlquery import WOQLQuery as WOQL


# Connect to the TerminusDB server
local="http://localhost:6363"
client = WOQLClient(server_url=local)

client.connect(team="admin")
exist = client.has_database("star-wars-tut")
if exist: client.delete_database("star-wars-tut")
```

```python
if not exist: client.create_database("star-wars-tut")
# Set the database and branch
client.connect(db="star-wars-tut", team="admin", key="root")
# len(list(client.get_all_documents()))
```

```python
# Insert schema from the JSON file
##################################
import json

with open("star-wars.json", "r") as f:
    schema = json.load(f)
    
# Insert the schema into TerminusDB
client.insert_document(schema, graph_type="schema",commit_msg = "Adding Star-wars Schema")
```

```python
# Insert instances from the Turtle file
#######################################
from rdflib import Graph
try:
   g = Graph()
   g.parse("star-wars-dataset.ttl", format="turtle")


   turtle_data = g.serialize(format='turtle')


   client.insert_triples("instance", turtle_data, "Inserting Star Wars data")
   client.commit()
  
except Exception as e:
   print(f"An error occurred: {e}")

# Print all triples just inserted
print (client.get_triples("instance"))
```
